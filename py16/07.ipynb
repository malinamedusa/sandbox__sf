```python
import json
import pandas as pd

with open('recipes.json') as f:
    recipes = json.load(f)
df = pd.DataFrame(recipes)
df.head()
```

```python

def get_data_frame(parent_column):

    columns = set()

    for index in range(df[parent_column].shape[0]):
        child_columns_name = df[parent_column][index]
        for column in child_columns_name:
            columns.add(column)

    print(len(list(columns)))

    for value in columns:
        df[value] = df[parent_column].apply(lambda x: 1 if value in x else 0)

    df[parent_column] = df[parent_column].apply(lambda x: len(x) if x != [] else 0)


get_data_frame('ingredients')
df
```

```python
df = pd.read_csv('recipes.csv')

ids  = list(set(df['id']))
ingredients = df.columns[3:]

def make_list(df):
    ingredients_true = []
    for i in ingredients:
        if df[i].item() == 1:
            ingredients_true.append(i)
    return  list(ingredients_true)

new_recipes = []

for current_id in ids:
    cuisine = df[df['id'] == current_id]['cuisine'].iloc[0]
    current_ingredients = make_list(df[df['id'] == current_id])
    current_recipe = {'cuisine': cuisine, 'id': int(current_id), 'ingredients': current_ingredients}
    new_recipes.append(current_recipe)
print(type(new_recipes))
```

```python
import json
new_recipes = json.dumps(new_recipes) # Функция dumps() модуля json сериализирует объект Python в строку формата JSON.
with open("new_recipes.json", "w") as write_file:
    write_file.write(new_recipes)
```
